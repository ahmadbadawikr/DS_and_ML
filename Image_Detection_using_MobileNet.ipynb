```
import keras
import numpy as np
from keras.preprocessing import image
import google.colab
from keras.applications import imagenet_utils
```

```
model = keras.applications.mobilenet.MobileNet()
```

```
model.summary()
```

```
fl = google.colab.files.upload() #memunculkan event button upload
img = image.load_img(open(next(iter(fl)),'rb'), target_size=(224,224))
imgarray = image.img_to_array(img)
```

```
imgarray = imgarray.reshape(1, 224, 224, 3)
imgarray = keras.applications.mobilenet.preprocess_input(imgarray)
```

```
ypred = model.predict(imgarray)
imagenet_utils.decode_predictions(ypred)
```
